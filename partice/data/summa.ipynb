CNN(Convolutional Neural Network, 합성곱 신경망)**은 주로 이미지 처리와 컴퓨터 비전 분야에서 사용되는 딥러닝 모델입니다. CNN은 이미지, 비디오, 시각적 데이터를 다룰 때 뛰어난 성능을 발휘하며, 현재 딥러닝 분야에서 가장 널리 사용되는 신경망 중 하나입니다.

CNN은 기본적으로 특성 추출과 분류 두 가지 주요 단계로 구성됩니다.

CNN의 주요 구성 요소
입력 레이어 (Input Layer):

CNN의 입력은 일반적으로 이미지입니다. 예를 들어, RGB 이미지는 높이 × 너비 × 채널 수(3채널)를 가진 3차원 배열로 입력됩니다. 각 픽셀 값은 0에서 255 사이의 값으로 표현됩니다.
합성곱 층 (Convolutional Layer):

CNN의 핵심은 합성곱(Convolution) 연산입니다. 합성곱 층은 필터(또는 커널)를 사용하여 이미지에서 **특징(Feature)**을 추출합니다. 이 필터는 이미지 위를 슬라이딩하며, 각 부분에 대해 필터와 이미지 패치의 점곱 연산을 수행합니다.

합성곱 연산을 통해 이미지에서 에지, 모서리, 질감 등의 중요한 정보를 추출할 수 있습니다.

예시:

만약 필터가 3x3 크기라면, 이미지의 3x3 부분에 적용되어 중요한 특성을 감지합니다. 필터의 크기, stride(필터가 이동하는 간격), padding(이미지 가장자리 처리 방법)에 따라 출력 크기가 결정됩니다.
활성화 함수 (Activation Function):

각 합성곱 연산 후에 비선형성을 추가하기 위해 활성화 함수가 적용됩니다. CNN에서 가장 많이 사용되는 활성화 함수는 **ReLU(Rectified Linear Unit)**입니다. ReLU는 음수를 0으로 바꾸고, 양수는 그대로 출력하는 함수입니다. 이를 통해 CNN은 복잡한 패턴을 학습할 수 있습니다.
풀링 층 (Pooling Layer):

풀링은 이미지의 크기를 줄이고 계산량을 감소시키며, 불필요한 정보를 줄이기 위한 작업입니다. **최대 풀링(Max Pooling)**과 **평균 풀링(Average Pooling)**이 주로 사용됩니다.
최대 풀링(Max Pooling): 일정 영역에서 가장 큰 값을 선택하여 출력합니다.
평균 풀링(Average Pooling): 일정 영역의 평균 값을 계산하여 출력합니다.
완전 연결 층 (Fully Connected Layer, Dense Layer):

CNN의 마지막 단계에서, 합성곱 층과 풀링 층에서 추출한 특징들을 바탕으로 분류를 수행합니다. 이 단계에서는 완전 연결 층(Dense Layer)을 사용하여 최종 분류 결과를 출력합니다.
이 층은 일반적인 신경망의 구성 요소와 동일하게 작동하며, 각 노드가 이전 층의 모든 노드와 연결됩니다.
출력층 (Output Layer):

출력층에서는 최종적으로 클래스 확률을 계산하여 이미지가 어떤 클래스에 속하는지 예측합니다. 이진 분류의 경우 시그모이드(Sigmoid) 함수가, 다중 클래스 분류의 경우 소프트맥스(Softmax) 함수가 사용됩니다.
CNN의 작동 원리 요약
입력: CNN은 입력 이미지로부터 시작됩니다.
합성곱: 필터(커널)를 사용하여 이미지에서 유용한 특징을 추출합니다.
활성화 함수: ReLU와 같은 활성화 함수를 통해 비선형성을 추가합니다.
풀링: 이미지의 차원을 줄여 계산량을 감소시키고, 중요한 정보를 유지합니다.
완전 연결 층: 추출된 특징을 기반으로 분류 작업을 수행합니다.
출력: 이미지가 어떤 클래스에 속하는지 확률을 출력합니다.
CNN의 장점
자동 특징 추출: CNN은 이미지에서 수동으로 특징을 추출할 필요 없이, 합성곱 연산을 통해 중요한 특징을 자동으로 학습합니다.
공간적 불변성: CNN은 이미지를 처리할 때 필터를 사용하여 공간적 구조를 유지하고, 위치에 상관없이 중요한 특징을 감지합니다.
효율적인 계산: 합성곱과 풀링을 통해 이미지의 크기를 줄이면서도 중요한 정보를 유지하여 효율적인 학습이 가능합니다.
CNN의 일반적인 사용 사례
이미지 분류: 이미지가 어느 클래스에 속하는지 예측 (예: 고양이 vs. 강아지).
객체 탐지(Object Detection): 이미지에서 객체의 위치를 찾아내고, 그것이 어떤 객체인지 분류.
이미지 생성: GAN(Generative Adversarial Network)와 같은 구조에서 이미지를 생성하거나 변형하는 데 사용.
영상 처리: 비디오 프레임의 특징을 학습하고 분석하는 데 사용.
의료 영상 분석: CT, MRI 스캔 등의 의료 이미지 분석에도 널리 사용.
CNN의 구조 예시

첫 번째 합성곱 층: 3x3 필터, 32개의 필터 사용.

출력 층: 10개의 클래스에 대한 확률(Softmax 사용).
결론
CNN은 이미지 인식 및 컴퓨터 비전 분야에서 매우 효과적인 딥러닝 모델입니다. 필터를 통해 자동으로 특징을 추출하고, 복잡한 이미지 패턴을 학습할 수 있으며, 다양한 이미지 관련 작업에서 뛰어난 성능을 발휘합니다.

CNN에서 **커널(kernel)**과 **풀 사이즈(pool size)**는 중요한 하이퍼파라미터입니다. 각각 데이터에서 특징을 추출하고, 차원을 줄이는 역할을 합니다. 여기서 kernel_size=3과 pool_size=2는 일반적인 선택이며, 그 의미를 이해하고 적절히 설정하는 것이 중요합니다.

1. 커널(kernel) 크기
**커널(또는 필터)**은 CNN의 합성곱 연산에서 이미지나 시계열 데이터의 특징을 추출하는 데 사용되는 작은 행렬입니다. 커널 크기는 CNN이 처리할 데이터의 작은 영역을 의미합니다.

kernel_size=3은 3개의 연속적인 데이터 포인트를 한 번에 처리한다는 뜻입니다. 이 값은 1차원 CNN에서 커널이 3개의 연속적인 특성을 사용해 합성곱 연산을 한다는 것을 의미합니다.
이유: 왜 kernel_size=3을 사용하는가?
작고 효율적인 커널: 3x3 크기의 커널은 CNN에서 많이 사용되며, 특히 이미지는 2D에서 3x3이, 1D 데이터에서는 3이 표준 크기로 많이 사용됩니다. 커널 크기가 작을수록 세부적인 특징을 잘 잡아낼 수 있습니다.
더 작은 크기는 세밀한 특징 추출에 적합: kernel_size=3은 더 작은 영역에서 세밀한 정보를 추출하는 데 적합합니다. 시계열 데이터에서 kernel_size=3을 사용하면 연속된 3개의 데이터 포인트를 묶어 특징을 추출할 수 있습니다.
2. 풀 사이즈(pool size)
**풀링(Pooling)**은 CNN에서 데이터를 다운샘플링하는 단계입니다. **풀 사이즈(pool size)**는 풀링 연산이 적용될 데이터 영역의 크기를 결정합니다. pool_size=2는 2개의 데이터 포인트를 하나로 줄인다는 의미입니다. 풀링은 데이터의 차원을 줄이면서 중요한 정보만 유지하는 역할을 합니다.

MaxPooling: 풀링 방식 중 가장 흔한 방법으로, pool_size 크기 안에서 최대값을 선택합니다.
이유: 왜 pool_size=2를 사용하는가?
차원 축소: 풀링은 데이터의 차원을 축소해 연산량을 줄이고, 모델이 학습하는데 필요한 자원을 줄이는 데 도움을 줍니다. pool_size=2는 데이터의 크기를 절반으로 줄여 연산량을 줄이면서도 중요한 정보를 유지하는 역할을 합니다.
과적합 방지: 풀링을 통해 차원을 줄임으로써 모델이 너무 많은 세부 사항에 집중하지 않도록 도와 과적합을 방지할 수 있습니다.
커널 크기 3과 풀 사이즈 2를 사용하는 이유
커널 크기 3: kernel_size=3은 입력 데이터에서 세밀한 특징을 추출할 수 있게 해줍니다. 특히 CNN에서 작은 커널은 더 깊은 층에서 더 복잡한 패턴을 학습할 수 있도록 도와줍니다.

풀 사이즈 2: pool_size=2는 합성곱 층에서 추출한 특징을 효율적으로 압축하고, 계산량을 줄이는 역할을 합니다. 데이터의 크기를 절반으로 줄이면서도 중요한 특징을 유지하는 좋은 방법입니다.

모델에서 이 값들이 어떻게 사용되었는지
Conv1D(32, kernel_size=3, activation='relu'): 이 합성곱 층은 32개의 필터를 사용해 1차원 입력 데이터에서 커널 크기 3으로 특징을 추출합니다. 즉, 입력 데이터에서 연속된 3개의 데이터 포인트를 묶어서 특징을 추출하고, 이를 통해 입력 데이터의 중요한 패턴을 찾아냅니다.

MaxPooling1D(pool_size=2): 이 풀링 층은 추출된 특징 맵에서 2개의 데이터 포인트 중 최대값을 선택하여, 차원을 절반으로 줄입니다. 이를 통해 연산량을 줄이고, 불필요한 정보를 줄이면서도 중요한 정보를 유지합니다.

결론
커널 크기(3): 작은 영역에서 세밀한 특징을 추출하는 데 사용됩니다. 1D CNN에서는 연속된 3개의 데이터 포인트를 처리하여 중요한 패턴을 학습합니다.
풀 사이즈(2): 차원을 줄이면서 중요한 정보를 유지하고, 모델이 더 일반화될 수 있도록 도와줍니다. pool_size=2는 데이터 크기를 절반으로 줄여 연산을 효율화하고, 과적합을 방지하는 역할을 합니다.
kernel_size=3과 pool_size=2는 CNN에서 자주 사용되는 표준 설정이며, 이 값들이 효율적이고 좋은 성능을 발휘하기 때문에 많은 모델에서 사용되고 있습니다.
